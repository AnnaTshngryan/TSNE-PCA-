```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
```

```python
X, y = load_digits(return_X_y=True)
```

```python
X.shape
```

```python
i = np.random.randint(0, X.shape[0])
plt.imshow(X[i].reshape(8, 8), cmap='Greys')
```

```python
#This step is not nessesary if we are using StandardScaler
X_normal = X/255
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X)
X_stand = scaler.transform(X)
```

```python
from sklearn.cluster import KMeans, MiniBatchKMeans
```

```python
mkmeans = MiniBatchKMeans(n_clusters=10)
mkmeans.fit(X_stand)
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_stand)
```

```python
tsne = TSNE(n_components=2, learning_rate=10)
X_tsne = tsne.fit_transform(X_stand)
```

```python
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='rainbow')
```

```python
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y, cmap='rainbow')
```

```python
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=mkmeans.labels_, cmap='rainbow')
```
